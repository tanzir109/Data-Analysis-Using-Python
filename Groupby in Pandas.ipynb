```python
import pandas as pd
```

```python
drinks = pd.read_csv(r"http://bit.ly/drinksbycountry")
```

```python
drinks.head(10)
```

```python
# mean of beer_serving across all the mentioned countries

drinks.beer_servings.mean()
```

```python
drinks.groupby('continent').beer_servings.mean()
```

```python
# How does it actually calculate on individual level?

drinks[drinks.continent == "Asia"].beer_servings.mean()
```

```python
drinks.groupby('continent').beer_servings.max()
```

```python
drinks.groupby('continent').beer_servings.agg(['count', 'mean', 'max', 'min'])
```

```python
# Grouping by multiple dimensions:

drinks.groupby(['continent', "country"]).beer_servings.agg(['count', 'mean', 'max', 'min'])
```
