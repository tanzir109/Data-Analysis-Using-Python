## Filling Null Values

```python
# You can press shift + tab to open intellisence

df_2 = df.fillna(value = 0)
df_2
```

```python
df_2.isnull().sum().sum()
```

```python
df.isnull().sum().sum()
```

```python
df_3 = df.fillna(value = 5)
df_3
```

```python
df_4 = df.fillna(method = 'pad')
df_4
```

```python
df_4.isnull().sum().sum()
```

## Filling null values with next value, row wise

```python
df_5 = df.fillna(method = "bfill")
df_5
```

## Filling null values with previous value, column wise

```python
df_6 = df.fillna(method = 'pad', axis = 1)
df_6
```

## Filling null values with next value, column wise

```python
df_7 = df.fillna(method = 'bfill', axis = 1)
df_7
```

## Filling nulls with different values in different columns

```python
df_8 = df.fillna({'society' : "abcd", 'balcony' : "defg"})
df_8
```

## Filling null values with the mean of a column

```python
df_9 = df.fillna(value = df['balcony'].mean())
df_9
```

## Dropna

```python
df_10 = df.dropna()
df_10
```

```python
df_11 = df.dropna(how = 'all')
df_11
```

```python
df_11 = df.dropna(how = 'any')
df_11
```

## replace()

```python
import numpy as np
df_12 = df.replace(to_replace = np.nan, value = 87546)
df_12
```

```python
df_13 = df.replace(to_replace = 3.0, value = 5.0)
df_13
```

## Interpolate()

```python
df['balcony'] = df['balcony'].interpolate(method = 'linear')
df
```
