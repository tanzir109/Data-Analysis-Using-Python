```python
import pandas as pd
import numpy as np
df = pd.read_csv(r"C:\2021\Tanzir_Data Analyst\Data\Olympic athletes.csv")
```

```python
df.head()
```

```python
# What was the average height of Olympic athletes?
df.Height.mean()
```

```python
# What was the average height of Olympic athletes per year?
df.groupby('Year').Height.mean()
```

```python
# What was the average height of athletes, per year and per country?
# for each distinct year -----------categorical data
# for each distinct team ------------categorical data
# what was the average height? In the value field, we need to have numerical value. 

df.pivot_table(index = "Year", columns = "Team", values = "Height").dropna()
```

```python
df.pivot_table(index = 'Year', columns = 'Team', values = 'Height').dropna(thresh = 10) 

# thresh = 10, mean, need to have al least 10 non-na value in the row to be included. 
```

```python
df.pivot_table(index = "Year", columns = 'Sport', values = "Height").dropna(thresh = 30)

# By default, you will get average of the values. However, we can change it by 'aggfunc' function
```

```python
df.pivot_table(index = 'Year', columns = 'Sport', values = "Height", aggfunc = 'max').dropna(thresh = 30)
```

```python
df.pivot_table(index = 'Year', columns = "Sport", values = "Height", aggfunc = ['mean', 'max']).dropna(thresh = 60)
```

```python
# The above code can also be written as "pd.pivot_table" instead of "df.pivot_table":

pd.pivot_table(df, index = 'Year', columns = "Sport", values = "Height", 
               aggfunc = ['mean', 'max'], margins = True, margins_name = "Overall Average Height").dropna(thresh = 60)
```
