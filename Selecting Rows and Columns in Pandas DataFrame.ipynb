```python
import pandas as pd
```

```python
ufo = pd.read_csv("http://bit.ly/uforeports")
```

```python
ufo.head(5)
```

```python
# loc is for selecting column / row by lebels

ufo.loc[[0, 1, 2], :]
```

```python
ufo.loc[0:2, :]  # In case of "loc", range is inclusive as it is by label.
```

```python
ufo.loc[0:2]
```

```python
ufo.loc[:, ['City', "State"]] # "loc" is by label.
```

```python
ufo.loc[:, 'City': "State"] # Even with label, you can set a range.
```

```python
# You can get the same thing, like this way:
ufo.drop("Time", axis =1)
```

```python
ufo[ufo.City == "Oakland"]
```

```python
# Remember "loc" is by label
ufo.loc[ufo.City == "Oakland", :]
```

```python
ufo.loc[ufo.City == "Oakland", 'State']
```

```python
# iloc is for filtering based on the index position
# It is exclusive of the last mentioned index

ufo.iloc[:, 0:4]
```

```python
ufo[['City', "State"]]
```

```python
# The above code can also be written as (multiple label can be used as a list)
ufo.loc[:, ['City', "State"]]
```

```python
# It is by index though "loc" or "iloc" are not mentioned.

ufo[0:2]
```

```python
# The above code can also be written as 

ufo.iloc[0:2, :]
```

## Conditional Selection of Data

```python
new_df = df[df['Bonus %']>5] 
new_df
```

```python
# "loc" deals with label, label can be included as condition, string and list.
df.loc[df['Salary'] > 50000, :]
```

```python
df.loc[df['Salary']!=50000, :]
```

```python
df.loc[df['Salary']==50000, :]  # There is no record.
```

### isin() function

```python
options = ['Product', "Client Services"]
new_df = df[df['Team'].isin(options)]
new_df
```

```python
# This is how it works: first: df['Team'].......gives you individual value of team, then isin(), provides boolean, after 
# thatn based on the list of True boolean, it gives the dataframe.
```

```python
df['Team']
```

```python
df['Team'].isin(options)
```

```python
df[df['Team'].isin(options)]
```

```python
df.loc[df['Team'].isin(options)]   # here, df['Team'].isin(options) provides label(boolean) as required for loc function.
```

```python
# It can be written as:
df.loc[df['Team'].isin(options), :]
```

```python
df.loc[~df['Team'].isin(options)]   # "~" act as "not", meaning other values will be selected. 
```

```python
options = ['Product', "Client Services"]
new_df = df[~df['Team'].isin(options)]
new_df
```

```python
# Applying condition to create a dataframe of choice

options = ['Sales']

new_df = df[(df['Gender']=='Male') & (df['Team'].isin(options))]
new_df
```
